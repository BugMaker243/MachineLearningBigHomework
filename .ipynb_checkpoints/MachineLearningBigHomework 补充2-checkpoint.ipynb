<div style="font-size:16px">
    <li><strong>（4）使用SiLU激活函数的一个合适大小的神经网络</strong></li>
</div>

<div style="font-size:17px; line-height:1.5">
    这里再使用SiLU激活函数构建一个神经网络，原因如下：<br><br>
    - ReLU激活函数在x大于0时是线性的，在x小于0时输出为0。ReLU函数的优点是计算简单，能够缓解梯度消失问题。但是，ReLU函数的缺点是它在x小于0时梯度为0，这可能会导致神经元"死亡"，即在训练过程中永远不会被激活。<br><br>
    - Sigmoid函数的公式为f(x) = 1 / (1 + exp(-x))。Sigmoid函数的输出在0到1之间，它能够将任意实数映射到0到1之间，因此常常被用于二分类问题的输出层。但是，Sigmoid函数在输入的绝对值较大时梯度接近0，这会导致梯度消失问题，使得网络难以学习。<br><br>
    - SiLU函数也被称为Swish函数，其公式为f(x) = x * sigmoid(x)。SiLU函数是一种平滑函数，它在负无穷大到正无穷大的范围内都是可导的，这使得它能够缓解梯度消失问题。SiLU函数在实践中通常能够比ReLU函数得到更好的性能。<br><br>
    - 我做YOLO项目做的比较多，在查看YOLOv8的源代码的时候发现YOLOv8将SiLU激活函数作为了神经网络的默认函数，如果不对激活函数做修改，项目会默认使用SiLU函数，这表明经过UltraLytics团队验证，SiLU激活函数在神经网络中效果非常好。虽然YOLO是CV方面的模型，我们大作业的项目是一个二分类任务，但是我希望尝试一下SiLU激活函数看看他在这个任务上的表现怎么样。
</div>

<div style="font-size:16px">
    定义神经网络结构
</div>

```python
import torch.nn.functional as F

class mid_SiLUNet(nn.Module):
    def __init__(self):
        super(mid_SiLUNet, self).__init__()
        self.fc1 = nn.Linear(8, 120)
        self.fc2 = nn.Linear(120, 120)
        self.fc3 = nn.Linear(120, 120)
        self.fc4 = nn.Linear(120, 120)
        self.fc5 = nn.Linear(120, 2)

    def forward(self, x):
        x = F.silu(self.fc1(x))
        x = F.silu(self.fc2(x))
        x = F.silu(self.fc3(x))
        x = F.silu(self.fc4(x))
        x = self.fc5(x)
        return x
```

<div style="font-size:16px">
    实例化网络、定义损失函数和优化器
</div>

```python
silu_model = mid_SiLUNet()
silu_criterion = nn.CrossEntropyLoss()
silu_optimizer = torch.optim.Adam(silu_model.parameters(), lr=0.001)
```

<div style="font-size:16px">
    搭建训练网络开始训练，这里会打印每一轮的训练损失和验证损失，方便训练的时候观察，同时因为训练时间比较长，这样可以减少等待时候的焦虑。
</div>

```python
silu_train_losses = []
silu_val_losses = []
for epoch in range(100):  # 迭代次数
    train_loss = 0
    val_loss = 0
    silu_model.train()
    for inputs, labels in train_loader:
        silu_optimizer.zero_grad()
        outputs = silu_model(inputs)
        loss = silu_criterion(outputs, labels)
        loss.backward()
        silu_optimizer.step()
        train_loss += loss.item()
    else:
        silu_model.eval()
        with torch.no_grad():
            for inputs, labels in val_loader:
                outputs = silu_model(inputs)
                loss = silu_criterion(outputs, labels)
                val_loss += loss.item()
        train_loss /= len(train_loader)
        val_loss /= len(val_loader)
        silu_train_losses.append(train_loss)
        silu_val_losses.append(val_loss)
        print("Epoch: {}/100.. ".format(epoch+1),
              "Training Loss: {:.3f}.. ".format(train_loss),
              "Validation Loss: {:.3f}.. ".format(val_loss))
```

<div style="font-size:16px; line-height:1.5">
    这个模型的训练效果从每一轮的输出上来看非常好
</div>

<div style="font-size:16px">
    验证模型并输出正确率
</div>

```python
# 测试网络
silu_model.eval()
silu_y_test_pred = torch.tensor([], dtype=bool)
silu_y_test_prob = torch.tensor([], dtype=torch.float32)
correct = 0
total = 0
with torch.no_grad():
    for inputs, labels in test_loader:
        outputs = silu_model(inputs)
        _, predicted = torch.max(outputs.data, 1)
        silu_y_test_pred = torch.cat((silu_y_test_pred, predicted == labels), dim=0)
        silu_y_test_prob = torch.cat((silu_y_test_prob, torch.softmax(outputs, dim=1)[:, 1]), dim=0)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print('准确率: %d %%' % (100 * correct / total))
```

<div style="font-size:16px; line-height:1.5">
    模型的准确率为97%，这一效果非常出色，证明了SiLU激活函数的优越性
</div>

<div style="font-size:16px">
    绘制学习曲线
</div>

```python
# 绘制学习曲线
plt.figure()
plt.plot(silu_train_losses, label='Training loss')
plt.plot(silu_val_losses, label='Validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Learning Curve')
plt.legend()
plt.show()
```

<div style="font-size:16px; line-height:1.5">
    虽然模型准确率非常好，但是从学习曲线可以看出模型可能出现了一些过拟合的现象，可能需要减少模型的复杂度
</div>

```python
from sklearn.metrics import roc_curve, auc

# 计算AUC
fpr, tpr, _ = roc_curve(silu_y_test_pred.numpy(), silu_y_test_prob.numpy())
roc_auc = auc(fpr, tpr)

# 绘制ROC曲线
plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='ROC曲线 (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('假正例率')
plt.ylabel('真正例率')
plt.title('ROC')
plt.legend(loc="lower right")
plt.show()
```

<div style="font-size:16px; line-height:1.5">
    在ROC曲线方面，效果依然不佳，更加印证了上一个模型分析中对神经网络可能在本项目数据集上有系统系问题，不太适合这个数据的认识。
</div>
