```python
# 测试网络
silu_model.eval()
silu_y_test_pred = torch.tensor([], dtype=bool)
silu_y_test_prob = torch.tensor([], dtype=torch.float32)
correct = 0
total = 0
with torch.no_grad():
    for inputs, labels in test_loader:
        outputs = silu_model(inputs)
        _, predicted = torch.max(outputs.data, 1)
        silu_y_test_pred = torch.cat((silu_y_test_pred, predicted == labels), dim=0)
        silu_y_test_prob = torch.cat((silu_y_test_prob, torch.softmax(outputs, dim=1)[:, 1]), dim=0)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print('准确率: %d %%' % (100 * correct / total))
```
