#### **二. XGB**

<div style="font-size:16px; line-height:1.5">
XGBoost，全称eXtreme Gradient Boosting ，简称XGB，是GBDT算法的一种变种，是一种监督算法；它是boost算法的一种，也属于集成算法，是一种伸缩性强、便捷的可并行构建模型的Gradient Boosting算法。其高效地实现了GBDT算法并进行了算法和工程上的许多改进，被广泛应用在Kaggle竞赛及其他许多机器学习竞赛中。<br>
该算法的实现旨在提高计算时间和内存资源的效率。设计目标是充分利用可用资源来训练模型。一些关键的算法实现特性包括：<br>

<li>&emsp;&emsp;具有自动处理缺失数据值的稀疏感知实现。<br>
<li>&emsp;&emsp;支持树构建并行化的块结构。<br>
<li>&emsp;&emsp;继续训练，以便您可以在新数据上进一步提升已经拟合的模型。<br><br>
下面我们使用该模型对本数据集进行训练
</div>

<div style="font-size:18px; line-height:1.5">
<li>boost:基分类器的烈性<br>
<li>num_class:几分类，这里是9<br>
<li>verbosity：打印消息的详细程度<br>

<li>scale_pos_weight：正样本的权重，在二分类模型中，如果两个分类的样本比例失衡，可以设置该参数，模型效果会更高。一般取值:负样本数/正样本数，一般模型里偏少的部分会认为是正样本。<br>
<li>max_depth: 树的深度，默认为6，一般取值[3,10] 越大偏差越小，方差越大，需综合考虑时间及拟合性。<br>
<li>min_child_weight：分裂叶子节点中样本权重和的最小值，如果新分裂的节点的样本权重和小于min_child_weight则停止分裂，默认为1，取值范围[0,],当值越大时，越容易欠拟合，当值越小时，越容易过拟合。<br>
<li>gamma:别名min_split_loss 制定节点分裂所需的最小损失很熟下降值,节点分裂时，只有损失函数的减小值大于等于gamma，节点才会分裂，gamma越大，算法越保守，取值范围为[0,] 【0,1,5】<br>
<li>subsample:训练每棵树时子采样比例，默认为1，一般在[0.5,1]之间，调节该参数可以防止过拟合。<br>
<li>colsample_bytree:训练每棵树时，使用特征占全部特征的比例，默认为1，典型值为[0.5,1]，调节该参数可以防止过拟合<br>
<li>alpha:别名reg_alpha，L1正则化，在高维度的情况下，调节该参数可以加快算法的速度，增大该值将是模型更保守，一般我们做特征选择的时候会用L1正则项<br>
<li>lambda:L2正则化，调节、、增大该参数可以减少过拟合，默认值为1<br>
<li>eta:别名learning_rate 学习率一般越小越好，只是耗时会更长<br>
<li>n_estimators:基学习器的个数，越大越好，偏差越小，但耗时会增加<br>
<li>max_delat_step:限制每棵树权重改变的最大步长，默认值为0，及没有约束，如果为正值，则这个算法更加保守，通常不需要设置该参数，但是当样本十分不平衡时，对逻辑回归有帮助。<br>
<li>nthread:有多少处理器可以使用，默认为1，-1表示没有限制。<br>
<li>silent:默认为0，不输出中间过程，=1，输出中间过程<br>
<li>seed:随机种子<br>

#### **三. AdaBoost**

<div style="font-size:16px; line-height:1.5">
AdaBoost（Adaptive Boosting）是一种经典的集成学习算法，其思想是将多个弱分类器（weak classifier）组合成一个强分类器（strong classifier），以提高分类性能。它是由 Yoav Freund 和 Robert Schapire 在1996年提出的。<br>

AdaBoost算法的基本思想是针对一个训练数据集，使用不同的分类算法（如决策树、神经网络、支持向量机等）构建一系列弱分类器，然后通过加权组合这些弱分类器，形成一个强分类器。
</div>

<div style="font-size:16px; line-height:1.5">
下面我根据Ada库以及查找来的资料构建了一个简单的Ada分类器
</div>

```python
class AdaBoostClassifier:
    def __init__(self, n_estimators=50, learning_rate=1.0):
        self.n_estimators = n_estimators  # 弱分类器个数
        self.learning_rate = learning_rate  # 学习率
        self.alpha = np.zeros(n_estimators)  # 弱分类器权重
        self.classifiers = []  # 弱分类器集合

    def fit(self, X, y):
        m, n = X.shape
        w = np.full(m, 1 / m)  # 初始化样本权重
        for t in range(self.n_estimators):
            # 使用决策树作为弱分类器
            clf = DecisionTreeClassifier(max_depth=1, random_state=2024)
            clf.fit(X, y, sample_weight=w)  # 基于样本权重训练弱分类器
            y_pred = clf.predict(X)  # 弱分类器对训练集的预测结果
            epsilon = np.sum(w[y_pred != y]) / np.sum(w)  # 计算加权错误率
            if epsilon > 0.5:
                break  # 如果加权错误率大于0.5，退出迭代
            self.alpha[t] = self.learning_rate * np.log((1 - epsilon) / epsilon)  # 计算弱分类器权重
            w *= np.exp(-self.alpha[t] * y * y_pred)  # 更新样本权重
            w /= np.sum(w)  # 样本权重归一化
            self.classifiers.append(clf)  # 添加弱分类器到集合中

    def predict(self, X):
        y_pred = np.zeros(len(X))
        for alpha, clf in zip(self.alpha, self.classifiers):
            y_pred += alpha * clf.predict(X)  # 弱分类器加权求和
        return np.sign(y_pred)  # 符号函数，判断正负号并返回预测标签

    # 计算准确率
    def score(self, y_pred, y):
        accuracy = (y_pred == y).sum() / len(y)
        return accuracy
```

<div style="font-size:16px; line-height:1.5">
上述类中函数的作用以及实现方法：<br>
&emsp;&emsp;（1）fit函数：训练函数。训练多个决策树基分类器，然后最终对这些基分类器进行加权得到最终输出结果。实现直接利用了 sklearn 中的 DecisionTreeClassifier，方便快捷。<br>
&emsp;&emsp;（2）predict函数：用于模型预测。使用测试数据集来验证训练出的AdaBoost模型的分类性能。对于测试数据集中的每个数据样本，利用AdaBoost模型进行预测，统计预测结果与真实标签相同的比例作为模型的准确率。<br>
&emsp;&emsp;（3）score函数：得出模型分数。由于AdaBoost是分类算法（也可以通过稍加改变实现回归），所以最终的衡量指标通常为准确率，所以我们需要计算出样本中分类正确的样本，然后将其除以总样本数得到准确率，<br>
</div>

<div style="font-size:16px; line-height:1.5">
分类器构建好了，下面进行训练
</div>

```python
model_ada = AdaBoostClassifier(n_estimators=100, learning_rate=1)
model_ada.fit(X_train, y_train)

y_train_pred = model_ada.predict(X_train)
y_test_pred = model_ada.predict(X_test)

score_train = model_ada.score(y_train_pred, y_train)
score_test = model_ada.score(y_test_pred, y_test)
```

#### **模型对比总结**

<div style="font-size:17px; line-height:1.5">
LGBM（LightGBM）和XGB（XGBoost）是基于梯度提升算法的模型，通常在分类问题上表现良好，特别是在处理复杂数据集和高维特征时效果显著。这些模型通过集成多个弱分类器来提升整体性能，并且能够有效地处理特征之间的复杂关系。在此问题中，则两个模型也确实表现出了良好的分类能力。<br>

相比之下，AdaBoost模型在此问题中表现不佳的原因可能包括：<br>

&emsp;&emsp;1. 对噪声敏感：AdaBoost模型对噪声和异常值比较敏感，这可能导致模型过拟合训练数据，从而影响模型在测试数据上的泛化能力。<br>

&emsp;&emsp;2. 数据不平衡：如果数据集存在严重的类别不平衡问题，AdaBoost模型可能会偏向于学习主要类别，而忽略少数类别，导致分类效果不佳。<br>

&emsp;&emsp;3. 弱分类器选择不当：AdaBoost模型的性能取决于弱分类器的选择，如果弱分类器过于简单或者过于复杂，都可能影响整体模型的性能。<br>

&emsp;&emsp;4. 学习率过高：过高的学习率可能导致模型在训练过程中过度拟合训练数据，从而影响模型的泛化能力。<br>
</div>
